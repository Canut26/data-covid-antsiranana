**COVID-19 Antsiranana, Madagascar**

```python
import sys
sys.path.insert(0, "../scripts")
from data_cleaning import clean_pipeline
from visualizations import generate_all
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = clean_pipeline()
print(f"{df.shape[0]} patients, {df.shape[1]} variables")
```

## 1. Profil démographique

```python
print(f"Âge médian : {df['Age_years'].median():.1f} ans")
print(f"Hommes : {(df['Sex']=='M').mean()*100:.1f}%")
print(f"Mortalité : {df['Outcome_binary'].mean()*100:.1f}%")
df[["Age_years","Sex","COVID_Severity","Outcome"]].describe(include="all")
```

## 2. Comorbidités

```python
comorbidities = ["Hypertension","Diabetes_Mellitus","Obesity","Heart_Disease","COPD","Chronic_Kidney_Disease"]
prevalence = df[comorbidities].mean()*100
prevalence.sort_values(ascending=False).plot(kind="barh", color="steelblue", figsize=(8,4))
plt.xlabel("Prévalence (%)")
plt.title("Prévalence des comorbidités")
plt.tight_layout()
plt.show()
```

## 3. Distribution de la sévérité

```python
df["COVID_Severity"].value_counts().plot(kind="bar", color=["#4CAF50","#FF9800","#F44336"])
plt.xticks(rotation=0)
plt.title("Distribution de la sévérité COVID-19")
plt.ylabel("Nombre de patients")
plt.tight_layout()
plt.show()
```
